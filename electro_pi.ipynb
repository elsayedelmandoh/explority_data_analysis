+ we show 3 lines in box:

  - the second line is `Q2` that median values for 50% of the population
  - the third line is `Q3` that median values for 75% of the population.

+ inter quarter range `(IQR) = Q3 - Q1`

+ two lines out of the box called `withcars`
  - `low` withcars = `Q1 - 1.5 * IQR` , values less than this value will be `Outliears`!
  - `high` withcars = `Q3 + 1.5 * IQR` , values higher than this value will be `Outliears`!
